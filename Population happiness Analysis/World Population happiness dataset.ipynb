##### Interpretation: 
##### With count, we see that there's no missing values and represents the number of observations availlable for analysis.
##### With mean, we see the central tendancy of the dataset.
##### With std, we have information about the variability or the degree of deviation from the mean.
##### The min and max  show us respectively the smallest and the largest values observed in each column.
##### 25%, 50%(median), 75% are the quartiles. The 25% (first quartile) representsthe value below wich 25% of data falls...

```python
# Checking for correlation between variables
data.corr(numeric_only = True)
```

```python
#Let's visualize correlation with a heatmap
corr_matrix = data.corr(method = 'pearson', numeric_only = True)
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_matrix, annot = True, fmt ='.2g', vmin=-1, vmax=1)
heatmap.set_title('Correlation')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

##### Interpretation: 
#####  Happiness is highly correlated to GDP per capita (0.78), ie the higher the gdp is , the higher is the score. It's quite the same with Life expectancy. However, there's nothing to do with generosity. Life expectancy is also highly correlated with GDP...

```python
# Scatter plot: Happiness Score vs. Economy GDP per Capita
sns.regplot(data = data, y = 'Happiness Score', x = 'Economy (GDP per Capita)', color = 'c').set(title = 'Happiness score vs Economy (GDP per Capita)')
```

### Task 2 : Answering some questions for better analysis and comprehension:

#### Question 1: What are the Top 10 most/least happy countries ?

```python
# Top 10 happiest countries
top_10_happiest_countries = data.sort_values('Happiness Score', ascending = False).head(10)
top_10_happiest_countries
```

```python
# Top 10 least happy countries
top_10_least_happy_countries = data.sort_values('Happiness Score', ascending = True).head(10)
top_10_least_happy_countries
```

#### Question 2: How does happiness vary across regions ? 

```python
data.groupby('Region')['Happiness Score'].mean(numeric_only = True).sort_values(ascending = False)
```

```python
data.groupby('Region')['Happiness Score'].mean(numeric_only = True).plot(kind = 'bar')
plt.xlabel('Region')
plt.ylabel('Happiness score')
plt.title('Average happiness score variation across regions ')
```

##### Interpretation:  The Australia and New Zealand Region has the highest happiness score and Sub-saharan Africa has the lowest

##### 

#### Question 3: What is the overall distribution of happiness scores among different countries? (Top 20)

```python
data.groupby('Country')['Happiness Score'].mean(numeric_only = True).sort_values(ascending = False)
```

```python
data.groupby('Country')['Happiness Score'].mean(numeric_only = True).head(20).plot(kind = 'barh')
plt.xlabel('Country')
plt.ylabel('Happiness score')
plt.title('Average happiness score variation across countries ')
```

#### Question 4: Which country is the most/least generous?

```python
most_generous_country = data.loc[data['Generosity'].idxmax()]
least_generous_country = data.loc[data['Generosity'].idxmin()]
print(f"The most generous country is: {most_generous_country['Country']} - {most_generous_country['Generosity']}")
print(f"The least generous country is: {least_generous_country['Country']} - {least_generous_country['Generosity']}")
```

#### Question 5: Does the level of social support (measured by the "Family" column) affect a country's happiness score? How does it compare to other factors like GDP per capita?

```python
# Scatter plot: Happiness Score vs.Family
sns.regplot(data = data, y = 'Family', x = 'Happiness Score', color = 'c').set(title = 'Family vs Happiness score')
```

#####  Since correlation between Family and Happiness score is high (0.74) and based on the scatter plot, we can say that the social support affects a country's happiness because the higher the support is the happier the country is.

```python
# Scatter plot: Family vs Economy (GDP per Capita)
sns.regplot(data = data, y = 'Family', x = 'Economy (GDP per Capita)', color = 'r').set(title = 'Family vs Economy (GDP per Capita)')
```

#####  Correlation between Family and Economy (GDP per Capita) is quite high (0.65) and based on the scatter plot, we can say that the Economy (GDP per Capita) affects a country's support. However, compared to the Happiness Score, the impact is lower.

#### Question 6: Is there a relationship between freedom and happiness?Are countries with higher levels of freedom more likely to have higher happiness scores?

```python
# Scatter plot: Freedom vs Happiness Score
sns.regplot(data = data, y = 'Freedom', x = 'Happiness Score', color = 'b').set(title = 'Freedom vs Happiness Score')
```
