##### We see that different population variables (2022 population, 2020 population... ) are highly correlated to each other  because over years population is growing in the same way. But with the other variables like area the correlation is low 

```python
# Let's visualize correlation with a heatmap
corr_matrix = data.corr(method = 'pearson', numeric_only = True)
sns.set(font_scale=0.65)
sns.heatmap(corr_matrix, annot = True, fmt = '.2g', vmin=-1, vmax=1, center=0, linewidths=0.5, linecolor='Black')
plt.figure(figsize = (14,6))
plt
```

##### We see that correlation between rank and other variables are negative and low, ie Rank does not have anything to do with other variables and the results. So as area, density, and growth rate. In the other hand population variables are highly correlated. 

### Grouping data

```python
#Let's look at population growth by continent

data.groupby('Continent').mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
```

##### Asia has the largest population in the world
